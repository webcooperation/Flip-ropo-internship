```python
# Question 1 : # python program to display all header tag from wikipedia
```

```python
url='https://en.wikipedia.org/wiki/Main_Page'
response = requests.get(url)
```

```python
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
# Find all header tags (h1 to h6) on the page
header_tags = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
# Extract the text content from the header tags
header_texts = [tag.get_text() for tag in header_tags]
header_texts
```

```python
# Create a data frame from the extracted header tags
import pandas as pd
df = pd.DataFrame({'Header': header_texts})
print(df)
```

```python
# Question 2 :  Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice)
```

```python
# Send a GET request to the website
url = "https://presidentofindia.nic.in/former-presidents.htm"
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the former presidents' information
table = soup.find("div", class_="innerContent alignLeft")

# Initialize empty lists to store the data
names = []
terms_of_office = []

# Extract the data from the table
rows = table.find_all("div",class_="presidentListing")  # Exclude the header row

for row in rows:
    #cells = row.find_all("td")
    name=row.find_all("h3")
    terms=row.find_all("p")
    #name = cells[0].text.strip()
    #term_of_office = cells[1].text.strip()
    names.append(name)
    terms_of_office.append(terms)

# Create a data frame from the extracted data
data = {
    "Name": names,
    "Term of Office": terms_of_office
}
df = pd.DataFrame(data)

# Print the data frame
print(df)
```

```python
 # 3a Write a python program to scrape cricket rankings from icc-cricket.com 
```

```python
# Send a GET request to the website
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the team rankings
table = soup.find("table", class_="table")

# Initialize empty lists to store the data
teams = []
matches = []
points = []
ratings = []

# Extract the data from the table
rows = table.find_all("tr")[1:11]  # Extract top 10 teams, excluding the header row
for row in rows:
    cells = row.find_all("td")
    team = cells[1].text.strip()
    match = cells[2].text.strip()
    point = cells[3].text.strip()
    rating = cells[4].text.strip()
    teams.append(team)
    matches.append(match)
    points.append(point)
    ratings.append(rating)

# Create a dataframe from the extracted data
data = {
    "Team": teams,
    "Matches": matches,
    "Points": points,
    "Rating": ratings
}

```

```python
df = pd.DataFrame(data)

# Print the dataframe
print(df)
```

```python
# Question 6 Write a python program to scrape the details of most downloaded articles from AI in last 90 days.
# https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below
# mentioned details and make data frame
```

```python
# Send a GET request to the website
url = "https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the container that holds the article details
container = soup.find("ul", class_="sc-9zxyh7-0 cMKaMj")

# Initialize empty lists to store the data
titles = []
authors = []
pdate = []

# Extract the data from the container
articles = container.find_all("li", class_="sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs")
for article in articles:
    title = article.find("a", class_="sc-5smygv-0 fIXTHm").text.strip()
    author = article.find("span", class_="sc-1w3fpd7-0 dnCnAO").text.strip()
    publisheddate = article.find("span",class_="sc-1thf9ly-2 dvggWt").text.strip()
    titles.append(title)
    authors.append(author)
    pdate.append(publisheddate)

# Create a data frame from the extracted data
data = {
    "Title": titles,
    "Author": authors,
    "Published date": pdate
}
df = pd.DataFrame(data)

# Print the data frame
print(df)
```
